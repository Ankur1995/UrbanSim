```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
from operator import itemgetter
```

```python
G = nx.read_graphml('london.graph.xml')
```

```python
print(nx.info(G))
```

```python
list(G.nodes(data = True))[0]
```

```python
df = nx.to_pandas_edgelist(G)
df[0:10]
```

```python
G_deg_1 = G.copy()
G_deg_2 = G.copy()

G_c_1 = G.copy()
G_c_2 = G.copy()

G_b_1 = G.copy()
G_b_2 = G.copy()
```

# Degree Centrality

```python
deg_london =nx.degree_centrality(G_deg_1)
nx.set_node_attributes(G_deg_1,dict(deg_london),'degree')
```

```python
df = pd.DataFrame()
df['station_name'] = G_deg_1.nodes()
df['degree'] = pd.Series(nx.get_node_attributes(G_deg_1, 'degree'))

df_sorted = df.sort_values(["degree"], ascending=False)
df_sorted = df_sorted.drop('station_name', axis=1)
df_sorted[0:10]
```

```python
sorted_ten= list(df_sorted.index)[0:10]
```

```python
def largest_component_ratio(graph):
    components = nx.connected_components(graph)
    largest = max(components, key=len)
    Largest_subgraph = graph.subgraph(largest)
    return (Largest_subgraph.number_of_nodes() / graph.number_of_nodes())
```

## Sequential

```python
for station in sorted_ten:
    G_deg_1.remove_node(station)
    print (nx.number_connected_components(G_deg_1))
    print(largest_component_ratio(G_deg_1))
```

## Non-Sequential

```python
for i in range(10):
    G_deg_2.remove_node(list(df_sorted.index)[0])

    deg_london =nx.degree_centrality(G_deg_2)
    nx.set_node_attributes(G_deg_2,dict(deg_london),'degree')

    df = pd.DataFrame()
    df['station_name'] = G_deg_2.nodes()
    df['degree'] = pd.Series(nx.get_node_attributes(G_deg_2, 'degree'))

    df_sorted = df.sort_values(["degree"], ascending=False)
    df_sorted = df_sorted.drop('station_name', axis=1)
    print('                                                                                        ')
    print(str(i+1))
    print(df_sorted[0:10])
    print(nx.number_connected_components(G_deg_2))
    print(largest_component_ratio(G_deg_2))
    print('xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx')
```

# Betweeness Centrality
